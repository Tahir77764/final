```python
import pandas as pd

# Load data
donors = pd.read_csv("bone_marrow_donor_2000_dataset_column_E_removed (1).csv")

# Normalize column names
donors.columns = donors.columns.str.strip().str.replace(" ", "_").str.lower()

# ---- SET CORRECT COLUMN NAMES HERE ----
HLA_COL = "hla"          # change if needed
AGE_COL = "age"
BLOOD_COL = "blood_group"
# --------------------------------------

# Check columns exist
for col in [HLA_COL, AGE_COL, BLOOD_COL]:
    if col not in donors.columns:
        raise ValueError(f"❌ Column '{col}' not found in dataset")

# Clean data
donors = donors.dropna(subset=[HLA_COL, AGE_COL, BLOOD_COL])
donors[AGE_COL] = donors[AGE_COL].astype(int)

def recommend_donors(hla_type, age, blood_group, top_n=5):
    filtered = donors[
        (donors[HLA_COL] == hla_type) &
        (donors[BLOOD_COL] == blood_group)
    ].copy()

    if filtered.empty:
        return "❌ No matching donors found"

    filtered["age_difference"] = abs(filtered[AGE_COL] - age)
    return filtered.sort_values("age_difference").head(top_n)

# Example test
recommend_donors("HLA-A*02", 28, "O+")
```

```python
import pandas as pd
import random

donors = pd.read_csv("bone_marrow_donor_2000_dataset_column_E_removed (1).csv")

# Normalize column names
donors.columns = donors.columns.str.strip().str.replace(" ", "_")

# Add HLA column if missing
if "HLA_Type" not in donors.columns:
    hla_types = ["HLA-A*01", "HLA-A*02", "HLA-B*07", "HLA-B*08", "HLA-DRB1*15"]
    donors["HLA_Type"] = [random.choice(hla_types) for _ in range(len(donors))]

# Clean data
donors = donors.dropna(subset=["Age", "Blood_Group", "HLA_Type"])
donors["Age"] = donors["Age"].astype(int)

def recommend_donors(hla_type, age, blood_group, top_n=5):
    filtered = donors[
        (donors["HLA_Type"] == hla_type) &
        (donors["Blood_Group"] == blood_group)
    ].copy()

    if filtered.empty:
        return "❌ No matching donors found"

    filtered["Age_Difference"] = abs(filtered["Age"] - age)
    return filtered.sort_values("Age_Difference").head(top_n)

# Test
recommend_donors("HLA-A*02", 28, "O+")
```
